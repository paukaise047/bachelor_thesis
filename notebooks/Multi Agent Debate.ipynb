```python

from autogen import GroupChat
from autogen import GroupChatManager
from autogen import ConversableAgent
import os

task = 'Please enter 7 words that are as different from each other as possible, in all meanings and uses of the words. Rules: 1. Only single words in English. 2. Only nouns (e.g., things, objects, concepts). 3. No proper nouns (e.g., no specific people or places). 4. No specialised vocabulary (e.g., no technical terms). 5. Think of the words on your own (e.g., do not just look at objects in your surroundings).'

def run_chat_instance(iteration):
    debater1 = ConversableAgent(
        "debater1",
        system_message= "You are a debater. Hello and welcome to the creativity competition, which will be conducted in a debate format. It 's not necessary to fully agree with each other 's perspectives, as our objective is to find the most creative list of words. The debate topic is stated as follows:" + task + "You can change words which are not different from each other, propose a new list. But remember to only have a list with 7 words. Not more.",
        llm_config={"config_list": [{"model": "gpt-3.5-turbo-0125", "temperature": 0.9, "api_key": "INSERT YOUR API KEY HERE"}]},
        human_input_mode="NEVER",  # Never ask for human input.
    )

    debater2 = ConversableAgent(
        "debater2",
        system_message= "You are a debater. Hello and welcome to the creativity competition, which will be conducted in a debate format. It 's not necessary to fully agree with each other 's perspectives, as our objective is to find the most creative list of words. The debate topic is stated as follows:" + task + "You can change words which are not different from each other, propose a new list. But remember to only have a list with 7 words. Not more.",
        llm_config={"config_list": [{"model": "gpt-3.5-turbo-0125", "temperature": 0.9, "api_key": "INSERT YOUR API KEY HERE"}]},
        human_input_mode="NEVER",  # Never ask for human input.
    )

    judge_message = "You are a judge." + task + "Your task is to decide which of the two list is better and ask then ask debater1 to continue making the chosen list more creative. Never stop asking to improve the list. You can give feedback. Tell them not to add to the list but iteratively change words."

    judge = ConversableAgent(
        "judge",
        system_message=judge_message,
        llm_config={"config_list": [{"model": "gpt-3.5-turbo-0125", "temperature": 0.9, "api_key": "INSERT YOUR API KEY HERE"}]},
        human_input_mode="NEVER",  # Never ask for human input.
    )

    group_chat = GroupChat(
        agents=[debater1, debater2, judge],
        messages=[],
        speaker_selection_method='round_robin',
        max_round=12,
    )

    group_chat_manager = GroupChatManager(
        groupchat=group_chat,
        llm_config={"config_list": [{"model": "gpt-3.5-turbo-0125", "api_key": "INSERT YOUR API KEY HERE"}]},
    )

    chat_result = judge.initiate_chat(
        group_chat_manager,
        message=task,
        summary_method="last_msg",
    )

for i in range(1, 31):
    print(f"Running chat instance {i}")
    run_chat_instance(i)
```
